## DATA EXPLORATORY AND VISUALISATION

```python
# New hotel booking dataset after the preparation
hotel.head()
```

```python
#Getting the preferred hotel based on the number of times it was booked
def count_rows(rows):
    return len(rows)
preferred_hotel = hotel.groupby(['Hotel Name']).apply(count_rows).reset_index(name='Count')
preferred_hotel_sorted= preferred_hotel.sort_values('Count', ascending=False)
print(preferred_hotel_sorted)
```

```python
#peferred hotel based on hotel rating
preferred_hotel = hotel.groupby(['Hotel Name','Hotel Rating']).size().reset_index(name='Count')
preferred_hotel_sorted= preferred_hotel.sort_values('Count', ascending=False)
print(preferred_hotel_sorted)
```

```python
#country most visited
def count_rows(rows):
    return len(rows)
most_visited_country = hotel.groupby(['Destination Country']).apply(count_rows).reset_index(name='Count')
most_visited_country_sorted= most_visited_country.sort_values('Count', ascending=False)
print(most_visited_country_sorted)
```

#### Most visted country based on the data is New Zealand

```python
def count_rows(rows):
    return len(rows)
pm = hotel.groupby(['Payment Mode']).apply(count_rows).reset_index(name='Count')
pms= pm.sort_values('Count', ascending=False)
print(pms)
```
